**What Is Data Science in Practice?**
Data Science is the collision of three skill sets. While some "unicorns" who are skilled in all three areas exist, most of us are a 2/3 stack engineers, who just google the rest of the area we don't know as much about. 

_

**What is Data Science as a Career?**
Two primary subfields: 
- Decision Science: Focused on providing predictive and prescriptive information to a business to make operational decisions. There is a stronger emphasis on maths and business in this subfield. Two examples: revenue forecasting, or identifying trends occurring in social media posts about a specific brand. 
- Applied Science: Focused on creating customer-facing application features powered by machine learning. Math and Computer Science are dominant knowledge domains in this space. Two examples: recommendation engines for driving consumer behavior, predictive text in email

## Data science

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. We'll go over this in depth today. For now: to edit the code, just click the cell and start editing.

You can either hit the play button in the upper left corner of the cell to run the code, or use the keyboard shortcut `shift`+`enter`. You'll see it output a visualization.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
